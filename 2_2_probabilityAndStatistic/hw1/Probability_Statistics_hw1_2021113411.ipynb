```python
from datascience import Table
import random
import pandas as pd
import numpy as np
path_data = '../../../'
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

random.seed(50);

head = 0;
num = [];
num_of_head = [];
proportion = [];
vialance = [];

for i in range(100):
    num.append(random.randrange(0, 2));
    if num[i] == 1:
        head += 1;
    num_of_head.append(head);
    proportion.append(head / (i + 1));
    vialance.append(proportion[i] * (1 - proportion[i]) / (i+1));

df = pd.DataFrame({'n' : np.arange(1, 101),'Head or Tail' : num, '# of head' : num_of_head, 'proportion' : proportion, 'vialance' : vialance});
table = Table.from_df(df);
table.show();
```

```python
table.plot('n', 'proportion');
```

```python
df_report = df.loc[(df.n %10 == 0) , ['n', 'proportion', 'vialance']];
table_report = Table.from_df(df_report);
table_report.show();
```
