```python
from datascience import *
%matplotlib inline
import random
import numpy as np
import pylab as pl
from scipy.stats import *
import matplotlib.pyplot as plots
```

```python
# 1.1 Binomial distritbution
# binomial initialization
np.random.seed(111)
binomial_n = 30
binomial_p = 0.5
binomial_repetition = 500

# binomial distribution
binomial_samples = np.random.binomial(binomial_n, binomial_p, 500)
binomial_Yi = binomial_samples / binomial_n             # binomial의 Yi

#print("binomial_Yi = ", binomial_Yi)

# binomial distribution graph
fig, ax1 = plots.subplots()
color_1 = 'tab:blue'
ax1.hist(binomial_Yi, color=color_1)

# ideal gaussian distribution
ideal_binomial_mean = binomial_p                                  # p_hat의 평균
ideal_binomial_var = binomial_p * (1 - binomial_p) / binomial_n   # p_hat의 분산
ideal_binomial_std = np.sqrt(ideal_binomial_var)                  # p_hat의 표준편차

print("ideal_binomial_mean = ", ideal_binomial_mean)
print("ideal_binomial_var = ", ideal_binomial_var)
print()

# ideal gaussian distribution graph
ax2 = ax1.twinx()
color_2 = 'tab:red'
x = np.linspace(ideal_binomial_mean - 4 * ideal_binomial_std, ideal_binomial_mean + 4 * ideal_binomial_std, 1000)
ax2.plot(x, norm.pdf(x, loc=ideal_binomial_mean, scale= ideal_binomial_std), color=color_2, linewidth=1.5)

fig.tight_layout()
plots.grid(False)
plots.show()
```

```python
# 1.2 Exponential distritbution
# exponential initialization
np.random.seed(111)
exponential_n = 30
exponential_lamda = 0.1
exponential_repetition = 500

# exponential distribution
exponential_samples = np.random.exponential(1/exponential_lamda, (exponential_repetition, exponential_n))
exponential_Yi = np.mean(exponential_samples, axis=1)   # exponential의 Yi

#print("exponential_Yi = ", exponential_Yi)

# exponential distribution graph
fig, ax1 = plots.subplots()
color_1 = 'tab:blue'
ax1.hist(exponential_Yi, color=color_1)

# ideal gaussian distribution
ideal_exponential_mean = 1 / exponential_lamda                          # x_bar의 평균
ideal_exponential_var = 1/ (exponential_lamda ** 2) / exponential_n     # x_bar의 분산
ideal_exponential_std = np.sqrt(ideal_exponential_var)                  # x_bar의 표준편차

print("ideal_exponential_mean = ", ideal_exponential_mean)
print("ideal_exponential_var = ", ideal_exponential_var)
print()

# ideal gaussian distribution graph
ax2 = ax1.twinx()
color_2 = 'tab:red'
x = np.linspace(ideal_exponential_mean - 4 * ideal_exponential_std, ideal_exponential_mean + 4 * ideal_exponential_std, 1000)
ax2.plot(x, norm.pdf(x, loc=ideal_exponential_mean, scale= ideal_exponential_std), color=color_2, linewidth=1.5)

fig.tight_layout()
plots.grid(False)
plots.show()
```

```python
# 1.3 Poisson distritbution
# poisson initialization
np.random.seed(111)
poisson_n = 30
poisson_lamda = 10
poisson_repetition = 500

# poisson distribution
poisson_samples = np.random.poisson(poisson_lamda, (poisson_repetition, poisson_n))
poisson_Yi = np.mean(poisson_samples, axis=1)   # poisson의 Yi

#print("poisson_Yi = ", poisson_Yi)

# poisson distribution graph
fig, ax1 = plots.subplots()
color_1 = 'tab:blue'
ax1.hist(poisson_Yi, color=color_1)

# ideal gaussian distribution
ideal_poisson_mean = poisson_lamda                              # x_bar의 평균
ideal_poisson_var = poisson_lamda / poisson_n                   # x_bar의 분산
ideal_poisson_std = np.sqrt(ideal_poisson_var)                  # x_bar의 표준편차

print("ideal_poisson_mean = ", ideal_poisson_mean)
print("ideal_poisson_var = ", ideal_poisson_var)
print()

# ideal gaussian distribution graph
ax2 = ax1.twinx()
color_2 = 'tab:red'
x = np.linspace(ideal_poisson_mean - 4 * ideal_poisson_std, ideal_poisson_mean + 4 * ideal_poisson_std, 1000)
ax2.plot(x, norm.pdf(x, loc=ideal_poisson_mean, scale= ideal_poisson_std), color=color_2, linewidth=1.5)

fig.tight_layout()
plots.grid(False)
plots.show()
```

```python
# 2 Normal distritbution
# normal initialization
np.random.seed(111)
normal_n = 30
normal_mean = 10
normal_variance = 3
normal_std = np.sqrt(normal_variance)
normal_repetition = 500
dof = normal_n - 1

# normal distribution
normal_samples = np.random.normal(normal_mean, normal_std, (normal_repetition, normal_n))
normal_Yi = np.mean(normal_samples, axis=1)
normal_Si2 = np.sum((normal_samples - normal_Yi[:, None])**2, axis=1) / dof

#print("normal_Si2 = ", normal_Si2)

# normal distribution graph
fig, ax1 = plots.subplots()
color_1 = 'tab:blue'
ax1.hist(normal_Si2, color=color_1)

# ideal gaussian distribution graph
ax2 = ax1.twinx()
color_2 = 'tab:red'
x = np.linspace(np.min(normal_Si2) - normal_std / 2, np.max(normal_Si2) + normal_std / 2, 1000)
ax2.plot(x, chi2.pdf(x, df = dof, scale = normal_variance / dof), color=color_2, linewidth=1.5) # (n-1)S^2 / var 이 자유도 n-1인 chi^2 분포를 따름 -> scale = var/dof

fig.tight_layout()
plots.grid(False)
plots.show()
```
